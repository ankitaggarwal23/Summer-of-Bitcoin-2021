```python
# Import pandas
import pandas as pd

# Import mempool.csv file into a Dataframe using pandas
df  = pd.read_csv("mempool.csv")

# Function to sort transaction 
def transaction_sort(df, max_fee, min_weight):
    df = df.sort_values([max_fee, min_weight], ascending = [False, True]).reset_index(drop = True)
    return df

# Function to check weight
def weight_check(A):
    if highest_weight >= minimum_weight + A['weight']:
        return True
    else:
        return False

# Function to check list
def list_check(A):
    if str(A) in final_txids:
        return True
    else:
        return False

# Function to check parent
def parent_check(A):
    if str(A[3]) != "nan":
        list_parent = str(A[3]).split(";")
        for idx in list_parent:
            if(list_check(idx)):
                continue
            else:
                txn_idx = df[df['tx_id'] == idx].index.item()
                B = df.loc[txn_idx]
                add_txn_check(B)

# Function to add weights to block 
def add_block(A):
    global minimum_weight
    txnID = A[0]
    weight = A[2]
    minimum_weight = minimum_weight + weight
    final_txids.append(txnID)

# Function to check transaction 
def add_txn_check(A):
    if(weight_check(A)):
        if(not list_check(A)):
            parent_check(A)
            if(weight_check(A)):
                add_block(A)

# Driver Function
def main(df):
    txn_sorted = transaction_sort(df, "fee", "weight")
    for idx in range(len(txn_sorted)):
        txnVar =  txn_sorted.loc[idx]
        add_txn_check(txnVar)

# Function to create block.txt file i.e generating output file
def write_to_file(final_list):
    file = open("block.txt","a")
    for idx in final_list:
        file.write(str(idx) + '\n')
    file.close()

# Initial Dataframe
highest_weight = 4000000
minimum_weight = 0
final_txids = []

data = df
main(data)

write_to_file(final_txids)
```
